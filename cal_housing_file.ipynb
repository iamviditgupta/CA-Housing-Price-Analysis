10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

**Using Num Py Lib method**

If the "total_bedrooms" is less than or equal to 10, it is categorized as "SMALL"

If the "total_bedrooms" is greater than 10 and less than or equal to 1000, it is categorized as "MEDIUM"

If the "total_bedrooms" is greater than 1000, it is categorized as "LARGE"

The np.select function is used to assign these categories based on the conditions. If none of the conditions are met, the value is set to "NAN" (Not A Number).







```python
c=[(hs["total_bedrooms"]<=10),                                             #condition list
  ( hs["total_bedrooms"]>10) & (hs["total_bedrooms"]<=1000),
   (hs["total_bedrooms"]>1000)]

v=["SMALL","MEDIUM","LARGE"]                                               #value list

hs["total_bedrooms_criteria"]=np.select(c,v,"NAN")


hs                                                                         #displaying dataset with size criteria of total_bedrooms
```
